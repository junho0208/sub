```python
word = input()
total = 0

for i in word :
    if 'a' <= i <= 'z' :
        total += ord(i) - 96
    else :
        total += ord(i) - 64 + 26

if total == 1 :
    print("It is a prime word.")
else :
    for i in range(2, total) :
        if total % i == 0 :
            print("It is not a prime word.")
            break
    else :
        print("It is a prime word.")
```
